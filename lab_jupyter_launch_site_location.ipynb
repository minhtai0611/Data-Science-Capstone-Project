Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
# Define a function to set marker color based on the class value
def set_marker_color(class_value):
    if class_value == 1:
        return 'green'
    else:
        return 'red'

# Apply the function to the class column to create the marker_color column
spacex_df['marker_color'] = spacex_df['class'].apply(set_marker_color)

# Display the last 10 rows to verify the new column
spacex_df.tail(10)
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Add marker_cluster to current site_map
site_map.add_child(marker_cluster)

# Iterate through each row in the dataframe
for index, record in spacex_df.iterrows():
    # Create a Marker object with its coordinates (Lat, Long)
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        # Customize the Marker's icon property based on the launch result
        icon=folium.Icon(color='white', icon_color=record['marker_color']),
        # Add a popup with the Launch Site and Class (Success or Failure)
        popup=f"Launch Site: {record['Launch Site']}, Class: {record['class']}"
    )
    # Add the marker to the marker cluster
    marker_cluster.add_child(marker)

# Add the marker cluster to the map
site_map.add_child(marker_cluster)

# Display the map
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
import numpy as np

# Function to calculate distance between two points on Earth using Haversine formula
def haversine(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    c = 2 * np.arcsin(np.sqrt(a))
    
    # Radius of Earth in kilometers
    r = 6371
    
    # Calculate the result
    return c * r
# Adding distances to the equator for each launch site
spacex_df['Distance to Equator (km)'] = spacex_df.apply(
    lambda row: haversine(row['Lat'], row['Long'], 0, row['Long']), axis=1
)

# Display the dataframe with the new distance column
spacex_df[['Launch Site', 'Lat', 'Long', 'Distance to Equator (km)']]
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
import folium
from folium.features import DivIcon

# Initialize the map centered on the launch site
launch_site_lat = 28.5623
launch_site_lon = -80.5774
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=10)

# Define coordinates for other important locations
railway_lat = 28.5701  # Example latitude for a railway
railway_lon = -80.6042  # Example longitude for a railway
highway_lat = 28.5634  # Example latitude for a highway
highway_lon = -80.5580  # Example longitude for a highway
city_lat = 28.5637  # Example latitude for a city
city_lon = -80.5716  # Example longitude for a city

# Create PolyLines between the launch site and these locations
def add_polyline(start_coords, end_coords, color='blue'):
    coordinates = [start_coords, end_coords]
    folium.PolyLine(locations=coordinates, weight=2, color=color).add_to(site_map)

# Add markers for each location
def add_marker(lat, lon, label, color='blue'):
    folium.Marker(
        location=[lat, lon],
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12px; color:{color};"><b>{label}</b></div>'
        )
    ).add_to(site_map)

# Add markers and lines
add_marker(launch_site_lat, launch_site_lon, 'Launch Site', color='blue')
add_marker(railway_lat, railway_lon, 'Railway', color='green')
add_marker(highway_lat, highway_lon, 'Highway', color='red')
add_marker(city_lat, city_lon, 'City', color='purple')

add_polyline([launch_site_lat, launch_site_lon], [railway_lat, railway_lon], color='green')
add_polyline([launch_site_lat, launch_site_lon], [highway_lat, highway_lon], color='red')
add_polyline([launch_site_lat, launch_site_lon], [city_lat, city_lon], color='purple')

# Display the map
site_map
```
